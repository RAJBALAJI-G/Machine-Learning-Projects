```python
from sklearn.preprocessing import LabelEncoder
```

```python
nonnumeric_columns = [df.columns[index] for index, dtype in enumerate(df.dtypes) if dtype == 'object']

for column in nonnumeric_columns:
    print(f"{column}: {df[column].unique()}")
```

```python
encoder = LabelEncoder()
for column in nonnumeric_columns:
    df[column] = encoder.fit_transform(df[column])
```

```python
newdf=df[['school','studytime','internet','failures','G1','G2','G3']]
```

```python
X=newdf.iloc[:,0:6].values
y=newdf.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(X_train, y_train)
```

```python
print(reg.score(X_train, y_train))
```

```python
y_pred=reg.predict(X_test)
```
